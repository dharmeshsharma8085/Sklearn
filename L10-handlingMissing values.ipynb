```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn.impute import SimpleImputer , KNNImputer

```

```python
data = {
    "Age":[25,np.nan ,30,28,np.nan],
    "Salaray":[50000 , 60000 , np.nan , 52000 , 50000]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Orginal data\n",df)
```

```python
simple_imputer = SimpleImputer(strategy = "mean")
df_simple = pd.DataFrame(simple_imputer.fit_transform(df) , columns=df.columns)
print("\nAfter SimpleImputer\n",df_simple )
```

Use median

```python
simple_imputer = SimpleImputer(strategy = "median")
df_simple1 = pd.DataFrame(simple_imputer.fit_transform(df) , columns=df.columns)
print("\nAfter SimpleImputer\n",df_simple1 )
```

Use constant

```python
simple_imputer = SimpleImputer(strategy = "constant")
df_simple2 = pd.DataFrame(simple_imputer.fit_transform(df) , columns=df.columns)
print("\nAfter SimpleImputer\n",df_simple2 )
```

```python
simple_imputer = SimpleImputer(strategy = "constant" , fill_value=10000)
df_simple2 = pd.DataFrame(simple_imputer.fit_transform(df) , columns=df.columns)
print("\nAfter SimpleImputer\n",df_simple2 )
```

```python
simple_imputer = SimpleImputer(strategy = "constant" , fill_value=10000)
df_simple2 = pd.DataFrame(simple_imputer.fit_transform(df[["Salaray"]]) , columns=["Salaray"])
print("\nAfter SimpleImputer\n",df_simple2 )
```

```python
#using KNN
```

```python
knn_imputer = KNNImputer(n_neighbors=2)
df_knn = pd.DataFrame(knn_imputer.fit_transform(df) , columns=df.columns)
print("\n After KNNImputer\n",df_knn)
```
